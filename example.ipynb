```python
import pandas as pd
from backtest import backtest as bt
from backtest import get_corr as corr
```

```python
# 计算20日换手率均值作为因子
turnover=pd.read_pickle("data/turnover.pickle")
factor=turnover.rolling(20,min_periods=10,axis=1).mean()
factor=factor.shift(2,axis=1)
```

```python
# 调用回测函数
bt(20130101,20211231,factor)
```

```python
corr(20130101,20211231,factor)
```
